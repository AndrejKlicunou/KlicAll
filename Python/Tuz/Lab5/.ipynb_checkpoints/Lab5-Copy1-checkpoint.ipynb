```python
import numpy as np
from collections import deque
from heapq import heappush, heappop
from matplotlib import pyplot as plt
from skimage.io import imshow, imshow_collection
import time
%pylab inline
```

```python
image2d = np.fromfile("circles.raw", dtype=np.int8).reshape(500, 500)
image3d = np.fromfile("id001.raw", dtype=np.int16).reshape(512, 512, 100)
```

```python
def gradient2d(image):
    axes = np.gradient(image)
    return np.power(axes[0] ** 2 + axes[1] ** 2, 0.5).astype(int)

def gradient3d(image):
    axes = np.gradient(image)
    return np.power(axes[0] ** 2 + axes[1] ** 2 + axes[2] ** 2, 0.5).astype(int)
```

```python
imshow(gradient2d(image2d))
```

```python
grad = gradient3d(image3d)
imshow_collection([grad[:,:,30], grad[:,50,:], grad[50,:,:]])
```

```python
def BeucherMeyer2dHeap(image, points):
    gradient = gradient2d(image)
    result = np.ones(image.shape).astype(np.int) * -1
    queue = []
    cur = 0
    for i in range(len(points)):
        result[points[i]] = i + 1
        heappush(queue, ((gradient[points[i]], cur), points[i]))
        cur += 1
    while len(queue) != 0:
        p = heappop(queue)[1]
        ways = []
        if p[0] != 0:
            ways.append((p[0]-1, p[1]))
        if p[0] != result.shape[0]-1:
            ways.append((p[0]+1, p[1]))
        if p[1] != 0:
            ways.append((p[0], p[1]-1))
        if p[1] != result.shape[1] - 1:
            ways.append((p[0], p[1]+1))
        for way in ways:
            if result[way] < 0:
                result[way] = result[p]
                heappush(queue, ((gradient[way], cur), way))
                cur += 1
    return result
    

def BeucherMeyer3dHeap(image, points):
    gradient = gradient3d(image)
    result = np.ones(image.shape).astype(np.int) * -1
    queue = []
    cur = 0
    for i in range(len(points)):
        result[points[i]] = i + 1
        heappush(queue, ((gradient[points[i]], cur), points[i]))
        cur += 1
    while len(queue) != 0:
        p = heappop(queue)[1]
        ways = []
        if p[0] != 0:
            ways.append((p[0]-1, p[1], p[2]))
        if p[0] != result.shape[0]-1:
            ways.append((p[0]+1, p[1], p[2]))
        if p[1] != 0:
            ways.append((p[0], p[1]-1, p[2]))
        if p[1] != result.shape[1] - 1:
            ways.append((p[0], p[1]+1, p[2]))
        if p[2] != 0:
            ways.append((p[0], p[1], p[2]-1))
        if p[2] != result.shape[2] - 1:
            ways.append((p[0], p[1], p[2]+1))
        for way in ways:
            if result[way] < 0:
                result[way] = result[p]
                heappush(queue, ((gradient[way], cur), way))
                cur += 1
    return result
```

```python
def BeucherMeyer2dHierarchical(image, points):
    gradient = gradient2d(image)
    result = np.ones(image.shape).astype(np.int) * -1
    queues = [deque() for i in range(np.max(gradient) + 1)]
    for i in range(len(points)):
        result[points[i]] = i + 1
        queues[gradient[points[i]]].append(points[i])
    for i in range(len(queues)):
            while len(queues[i]) != 0:
                p = queues[i].popleft()
                ways = []
                if p[0] != 0:
                    ways.append((p[0]-1, p[1]))
                if p[0] != result.shape[0]-1:
                    ways.append((p[0]+1, p[1]))
                if p[1] != 0:
                    ways.append((p[0], p[1]-1))
                if p[1] != result.shape[1] - 1:
                    ways.append((p[0], p[1]+1))
                for way in ways:
                    if result[way] < 0:
                        result[way] = result[p]
                        queues[max(gradient[way], i)].append(way)
    return result


def BeucherMeyer3dHierarchical(image, points):
    gradient = gradient3d(image)
    result = np.ones(image.shape).astype(np.int) * -1
    queues = [deque() for i in range(np.max(gradient) + 1)]
    for i in range(len(points)):
        result[points[i]] = i + 1
        queues[gradient[points[i]]].append(points[i])
    for i in range(len(queues)):
            while len(queues[i]) != 0:
                p = queues[i].popleft()
                ways = []
                if p[0] != 0:
                    ways.append((p[0]-1, p[1], p[2]))
                if p[0] != result.shape[0]-1:
                    ways.append((p[0]+1, p[1], p[2]))
                if p[1] != 0:
                    ways.append((p[0], p[1]-1, p[2]))
                if p[1] != result.shape[1] - 1:
                    ways.append((p[0], p[1]+1, p[2]))
                if p[2] != 0:
                    ways.append((p[0], p[1], p[2]-1))
                if p[2] != result.shape[2] - 1:
                    ways.append((p[0], p[1], p[2]+1))
                for way in ways:
                    if result[way] < 0:
                        result[way] = result[p]
                        queues[max(gradient[way], i)].append(way)
    return result
```

```python
points2d = [(100, 100), (250, 250), (350, 350), (480, 480)]
```

```python
start = time.time()
heap2d = BeucherMeyer2dHeap(image2d, points2d)
print('2D with heap: {0:.1f} sec'.format(time.time() - start))
imshow(heap2d, cmap=cm.viridis)
```

```python
start = time.time()
hierarchical2d = BeucherMeyer2dHierarchical(image2d, points2d)
print('2D with hierarchical PQ: {0:.1f} sec'.format(time.time() - start))
imshow(hierarchical2d, cmap=cm.viridis)
```

```python
points3d = [(100, 100, 40), (250, 250, 40), (350, 350, 40), (420, 420, 40)]
```

```python
start = time.time()
heap3d = BeucherMeyer3dHeap(image3d, points3d)
print('3D with heap: {0:.1f} sec'.format(time.time() - start))
```

```python
start = time.time()
hierarchical3d = BeucherMeyer3dHierarchical(image3d, points3d)
print('3D with hierarchical PQ: {0:.1f} sec'.format(time.time() - start))
```
