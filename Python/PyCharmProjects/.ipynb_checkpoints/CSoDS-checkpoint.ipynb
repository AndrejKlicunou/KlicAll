```python
import matplotlib.pylab as plt
import numpy as np

text_lengths_file = open('tl.txt', 'r')
text_lengths = np.array(text_lengths_file.readline().strip().split(' '))
tl_probas_file = open('tl_probas.txt', 'r')
tl_probas = np.array(tl_probas_file.readline().strip().split(' '))

key_lengths_file = open('kl.txt', 'r')
key_lengths = np.array(key_lengths_file.readline().strip().split(' '))
kl_probas_file = open('kl_probas.txt', 'r')
kl_probas = np.array(kl_probas_file.readline().strip().split(' '))


plt.plot(key_lengths, kl_probas, color = 'darkred', label = '')
plt.title('')
plt.xlabel('key length')
plt.ylabel('decode accuracy')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python

plt.plot(text_lengths, tl_probas, color = 'darkred', label = '')
plt.title('')
plt.xlabel('text length')
plt.ylabel('decode accuracy')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
