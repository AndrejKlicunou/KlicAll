```python
from scipy import fftpack
from PIL import Image
from scipy import misc
import numpy as np
from skimage import io
import matplotlib
from matplotlib import pyplot as plt
```

```python
image = misc.imread('saturn.jpg', 'P')
transform = fftpack.fftshift(fftpack.fft2(image))
ampl = np.log(1 + np.absolute(transform))
phase = np.angle(transform)
```

**Амплитудный и фазовый спектры:**

```python
io.imshow_collection([ampl, phase])
plt.show()
```

**По амплитудному спектру можно сказать, что изображение имеет много линий под углом около 45 градусов, т.к. на изображении спектра эти линии есть.  Ещё коэффициенты для низких частот относительно больше, что говорит о примерной однотонности изображения.**

**Изображения после удаления низких и высоких частот (в качестве низких частот будем брать прямоугольники из центра изображения размера $(i+1) \times (i+1)$, $i = \overline{1,5}$): изображение слева - после удаления низких частот, в центре - после удаления высоких, справа - исходное изображение.**

```python
for i in range(1, 6):
    without_small_freq = transform.copy()
    without_small_freq[128 - i: 128 + i, 128 - i : 128 + i] = 0
    with_small_freq = np.zeros([256, 256], dtype = complex)
    with_small_freq[128 - i:128 + i, 128 - i:128 + i] = transform[128 - i:128 + i, 128 - i:128 + i]
    large_freq = fftpack.ifft2(fftpack.ifftshift(without_small_freq))
    small_freq = fftpack.ifft2(fftpack.ifftshift(with_small_freq))
    inv_image = fftpack.ifft2(fftpack.ifftshift(transform))
    io.imshow_collection([np.absolute(large_freq), np.absolute(small_freq), np.absolute(inv_image)])
    plt.show()
```

**Задача поиска клада:**

```python
from skimage.measure import label
from skimage.color import label2rgb
from skimage.measure import regionprops
```

**Выделение стрелок и возможных кладов: стрелка содержит в себе от 1400 до 1600 пикселей. Определение красной стрелки: стрелка будет красной, если её центр является таковым.**

```python
def find_arrows(regionproperties):
    result = []
    for prop in regionproperties:
        if 1600 > prop.area > 1400:
            result.append(prop)
    return result


def find_treasures(regionproperties):
    result = []
    for prop in regionproperties:
        if not 1600 > prop.area > 1400:
            result.append(prop)
    return result


def find_red_arrow(arrows, image):
    for arrow in arrows:
        if 80 > image[int(arrow.centroid[0]), int(arrow.centroid[1])] > 73:
            return arrow
```

**Поиск объекта (компоненты), на который направлена текущая стрелка: для неё берётся угол между главной осью стрелки и осью Х, по нему определяем шаг, с которым будем двигаться в двух направлениях и смотреть, лежит ли текущая точка (по каждому из направлений) в каком-то прямоугольнике, ограничивающем некоторый объект. Если да, то проверяем, не является этот объект самим собой или тем объектом, который направляет стрелку на наш объект.**

```python
def find_next_object(arrow, prev_arrow, properties):
    first_dir = list(arrow.centroid)
    second_dir = list(arrow.centroid)
    step_x = 10 * np.cos(arrow.orientation)
    step_y = -10 * np.sin(arrow.orientation)
    while True:
        first_dir[0] += step_y
        first_dir[1] += step_x
        second_dir[0] -= step_y
        second_dir[1] -= step_x
        for prop in properties:
            rect = prop.bbox
            if rect[0] < first_dir[0] < rect[2] and \
               rect[1] < first_dir[1] < rect[3]:
                    if rect != arrow.bbox and (prev_arrow == None or rect != prev_arrow.bbox):
                        return prop
            if rect[0] < second_dir[0] < rect[2] and \
               rect[1] < second_dir[1] < rect[3]:
                    if rect != arrow.bbox and (prev_arrow == None or rect != prev_arrow.bbox):
                        return prop
```

**Функция, которая ищет клад среди выделенных объектов на изображении по начальной стрелке.**

```python
def find_treasure_comp(red_arrow, arrows, treasures):
    all_obj = arrows + treasures
    prev_obj = None
    cur_obj = red_arrow
    while True:
        prev_obj = find_next_object(cur_obj, prev_obj, all_obj)
        prev_obj, cur_obj = cur_obj, prev_obj
        for treasure in treasures:
            if cur_obj.bbox == treasure.bbox:
                return cur_obj
```

**Итоговая функция, которая на иозбражении ищет клад:**

**Функция для визуализации выделенных компонент в изображении:**

```python
def find_coloured_components(image):
    binarized = (images[i] > limit).astype(int)
    components  = label(binarized)
    return label2rgb(components)
```

**Эмпирически был подобрано значение порога, равное 32.**

```python
limit = 32
images = []
treasures = []
coloured_components = []
binarized = []
for i in range(3):
    images.append(misc.imread('Klad0' + str(i) + '.jpg', 'P'))
    treasures.append(find_treasure(images[i], limit))
    coloured_components.append(find_coloured_components(images[i]))
    binarized.append((images[i] > limit).astype(int))
```
