**Алгоритм разбивает 64-битный алгоритм на 4 равные части, по каждой из них строит 4 std::unordered_map, где ключ - 16-битная часть симхэша, а значение - множество симхэшей, содержащее в себе часть, являющуюся ключом. Это делает функция get_indices(). Далее выбираем первый по порядку в массиве симхэш, для которого не определена группа (делаем это, пока этот симхэш без группы существует), и просматривая 4 множества, которые соответствуют 16-битным частям этого симхэша, строим новую группу. После чего все вхождения симхэшэй группы удаляем из словарей.**

```python
file = open('result.txt', 'r')
data = [line.strip().split(' ') for line in file]
file.close()
```

```python
x = []
y = []
for item in data:
    x.append(int(item[0]))
    y.append(int(item[1]))
```

```python
import numpy as np
log_x = np.log(x)
log_y = np.log(y)
```

```python
plt.plot(log_x, log_y, color = 'blue', label = 'distribution curve')
plt.title("log distribution")
plt.xlabel('log(sizes)')
plt.ylabel('log(counts)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
