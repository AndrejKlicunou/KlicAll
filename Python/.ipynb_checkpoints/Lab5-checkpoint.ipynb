```python
import matplotlib.pylab as plt
%pylab inline
```

```python
import numpy as np
```

```python
Mh = np.array([[2, -2, 1, 1], 
               [-3, 3, -2, -1], 
               [0, 0, 1, 0], 
               [1, 0, 0, 0]])
```

```python
#inputting P1 = axis(0), P4 = axis(1), R1 = axis'(0), R4 = axis'(1) for x, y axes
val_x = [1, 1, 1, 1]
val_y = [2, -1, 1, 1]
val_z = [0, 1, 2, -2]
```

```python
c_x = Mh.dot(val_x)
c_y = Mh.dot(val_y)
c_z = Mh.dot(val_z)
```

```python
print(c_x)
print(c_y)
print(c_z)
```

```python
x_s = []
y_s = []
z_s = []
for t in np.arange(0, 1.001, 0.001):
    x_s.append(np.sum([c_x[i] * (t ** (3-i)) for i in range(4)]))
    y_s.append(np.sum([c_y[i] * (t ** (3-i)) for i in range(4)]))   
    z_s.append(np.sum([c_z[i] * (t ** (3-i)) for i in range(4)]))   
```

```python
plt.figure(figsize=(10, 10))
plt.plot(x_s, y_s)
plt.grid()
plt.title('Curve')
plt.ylabel('y')
plt.xlabel('x')
plt.legend(['(x(t), y(t))'], loc=2)
plt.show()
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x_s, y_s, z_s)

plt.show()
```
