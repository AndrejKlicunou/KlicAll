```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons, load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
X, y = make_moons(noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
forest = RandomForestClassifier(n_estimators=9, random_state=12, n_jobs=-1)
forest.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import mglearn
plt, axes = plt.subplots(3, 3, figsize=(30, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title(f'Tree {i}')
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1], alpha=0.4)
axes[-1, -1].set_title('Random forest')
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
forest = RandomForestClassifier(n_estimators=10, random_state=12, n_jobs=-1)
forest.fit(X_train, y_train)
```

```python
print(f'Train set accuracy {forest.score(X_train, y_train)}')
print(f'Test set accuracy {forest.score(X_test, y_test)}')
```
