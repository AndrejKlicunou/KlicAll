```python
import math
input_pair = [1.0, 0.0]
import random
```

```python
def loss(y, u):
    assert(len(y) == len(u))
    loss = 0.0
    for i in range(len(y)):
        loss += (u[i] - y[i]) * (u[i] - y[i])
    return loss / len(y)

def activate(data):
    return [1.0 / (1.0 + math.exp(-item)) for item in data]

def nn(input_data, residual):
    first_layer = [0.45, 0.78, -0.12, 0.13]
    
    first_layer_output = activate((input_data[0] * first_layer[0] + input_data[1] * first_layer[2], 
                          input_data[0] * first_layer[1] + input_data[1] * first_layer[3]))
    
    second_layer = [1.5, -2.3]
    
    result_before_sigm = first_layer_output[0] * second_layer[0] + first_layer_output[1] * second_layer[1] + residual
    
    result_after_sigm = 1.0 / (1.0 + math.exp(-result_before_sigm))
    return result_after_sigm
```

```python
baseline = (0, 1, 1, 0)
input_data = ([0, 0], [0, 1], [1, 0], [1, 1])
for res in [0.1 * i for i in range(10)]:
    data = []
    print('nn out: ', end='')
    for item in input_data:
        data.append(nn(item, res))
        print(data[-1], end=' ')
    print("loss: ", loss(baseline, data))
    print()
```
