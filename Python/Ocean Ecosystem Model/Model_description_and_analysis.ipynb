**Model description:** 

Ocean is a square field with cells. First state of ocean is generated by pseudo-randomized algorithm. Every cell has probability 55% to be a cell with water, 5% to be a cell with barrier (nobody may move at this cell), 25% to be a cell with victim and 15% to be a cell with predator. Model takes a size of squared ocean as a parameter.

Predators take such parameters as pregnancy term (after reaching it predator may reproduce) and hungriness limit (if predator reaches this limit, it will die). Primarily predators eat victims if possile, in the second place predators reproduce if they may and in the last place predators move randomly and equiprobably in any heighboring by side cell they may to move (current cell is a direction to 'move' too).

Victims take parameter as pregnancy term (after reaching it predator may reproduce). In the first place victims reproduce if they may and in the last place victims move randomly and equiprobably in any heighboring by side cell they may to move (current cell is a direction to 'move' too).

For reproducing animals have to have heighboring by side cell with water.

For eating predators have have to have heighboring by side cell with victim.

In model simulation in every iteration primarily predators perform their actions, then stayed alive victims perform actions. Simulation itertation will finish when all animals repform their actions.

Simulation will finish if all animals of any type are dead or simulation iterations count exceeds iterations limit which is taken as a parameter.

**Code for visualization:**

```python
import matplotlib.pylab as plt
import numpy as np

predators_file = open('predators.txt', 'r')
predators_counts = np.array(predators_file.readline().strip().split(' '))
victims_file = open('victims.txt', 'r')
victims_counts = np.array(victims_file.readline().strip().split(' '))
time = np.arange(victims_counts.shape[0])

plt.plot(time, predators_counts, color = 'darkred', label = 'predators')
plt.plot(time, victims_counts, color = 'darkgreen', label = 'victims')
plt.title('victims and predators')
plt.xlabel('time')
plt.ylabel('animal count')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

On plots we can see how animals' populations change by time. There we can see some time periods where appearance of populations looks like similarly.
