**Реализовал кодирование и декодирование, генерирую p для геометрического распределения по формуле $e ^{-k \alpha}$, где $\alpha$ равномерно распределена интервале $[0,1]$, $k$ из множества {10, 15}. С фиксированным p генерируется 10000 чисел, на которых и считается качество сжатия. Так делаю по 1000 раз с различным $k$, результаты, по которым строятся графики, записываю в log.txt. Вывожу 4 части, чтобы было наглядней видно, как значение p влияет на коэффициент качества сжатия.** 

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
file = open('log.txt', 'r')
data = []
for line in file:
    data.append(line.strip().split(' '))
    for i in range(len(data[-1])):
        data[-1][i] = float(data[-1][i])
```

```python
data = list(sorted(data))
```

```python
koeff = [item[1] / item[2] for item in data]
p = [item[0] for item in data]
```

```python
plt.figure(figsize = (20, 10))
plt.plot(p[:500], koeff[:500], color = 'darkblue', label = 'coeff(p)')
plt.title('quality of compression, p in [{}, {}]'.format(p[0], p[499]))
plt.legend(loc='upper left')
plt.xlabel('probability of geometric distribution')
plt.ylabel('compression coefficient')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
plt.plot(p[500:1000], koeff[500:1000], color = 'darkblue', label = 'coeff(p)')
plt.title('quality of compression, p in [{}, {}]'.format(p[500], p[999]))
plt.legend(loc='upper left')
plt.xlabel('probability of geometric distribution')
plt.ylabel('compression coefficient')
plt.show()
```
