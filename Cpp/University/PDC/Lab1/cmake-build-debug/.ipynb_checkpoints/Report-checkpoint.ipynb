```python
import matplotlib.pylab as plt
import numpy as np

files = [[], []]
for i in range(2):
    for j in range(12):
        file = open(str(i) + str(j) + '.txt', 'r')
        data = np.array(file.readline().strip().split(' '))
        files[i].append(data)

best_sizes = []
for i in range(2):
    file = open(str(i) + 'bb_sizes.txt', 'r')
    data = (file.readline().strip().split(' '))
    best_sizes.append(data)
    
```

```python
time = np.arange(files[0][0].shape[0])
plt.plot(time, files[0][0], color = 'darkred', label = 'Pointwise')
plt.plot(time, files[0][2], color = 'blue', label = 'Block')
plt.title('block_size = 1, N1 = 400, N2 = 400, N3 = 400')
plt.xlabel('iteration number')
plt.ylabel('execution time in milliseconds')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
plt.plot(time, files[0][1], color = 'darkgreen', label = 'Pointwise parallel')
plt.plot(time, files[0][3], color = 'yellow', label = 'Block parallel')
plt.title('block_size = 1, N1 = 400, N2 = 400, N3 = 400')
plt.xlabel('iteration number')
plt.ylabel('execution time in milliseconds')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
